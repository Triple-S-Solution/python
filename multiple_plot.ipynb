```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=1,ncols=3,sharex=True, sharey=True, figsize=(25,8))
ax[0].set_title('Grafik tanX')
ax[0].plot(x, tan, color='blue',label='np.tan(x)/10')
ax[1].set_title('Grafik sinX')
ax[1].plot(x, sin,color='red',label='np.sin(x)')
ax[2].set_title('Grafik cosX')
ax[2].plot(x, cos,color='green',label='np.cos(x)')

fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(sharex=True, sharey=True, figsize=(12,8))
ax.plot(x,y,color='blue')
a=ax.inset_axes([-0.2,-0.2,0.3,0.3])
b=ax.inset_axes([-0.2,0.9,0.3,0.3])
c=ax.inset_axes([0.9,-0.2,0.3,0.3])
d=ax.inset_axes([0.9,0.9,0.3,0.3])
a.set_title('Grafik cosX^2')
b.set_title('Grafik cosX^5')
c.set_title('Grafik cosX^2')
d.set_title('Grafik cosX^3')
a.plot(x,y2,color='red')
a.plot(x,y,color='blue')
b.plot(x,y5,color='yellow')
b.plot(x,y,color='blue')
c.plot(x,y3,color='green')
c.plot(x,y,color='blue')
d.plot(x,y4,color='black')
d.plot(x,y,color='blue')


plt.show()
```
